# Batch

```python
with open('batch_data.pkl', 'rb') as f:
    batch = pickle.load(f)
```

```python
print(len(batch))
```

```python
print(batch)
```

```python
print(batch.keys())
```

```python
print(batch["im_file"])
print(batch["ori_shape"])
print(batch["resized_shape"])
print(batch["ratio_pad"])
print()
print(batch["img"])
```

```python
import matplotlib.pyplot as plt
import torch

# Charger les images
images = batch['img']  # Tensor de forme [B, C, H, W]

# Boucle pour afficher chaque image
for i in range(images.size(0)):  # Iterer sur le batch
    img = images[i].cpu().numpy()  # Déplacer sur CPU et convertir en NumPy
    img = img.transpose(1, 2, 0)  # Réorganiser les dimensions de [C, H, W] à [H, W, C]
    
    # Les images YOLO sont souvent normalisées, alors denormaliser si nécessaire
    img = (img * 255).astype('uint8')  # Si normalisé entre 0 et 1
    
    # Afficher l'image
    plt.figure(figsize=(6, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Image {i}")
    plt.show()
```
